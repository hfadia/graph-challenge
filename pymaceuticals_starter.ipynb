```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st




# Study data files
mouse_metadata_path = "Mouse_metadata.csv"
study_results_path = "Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
pymaceutical_data_complete = pd.merge(study_results, mouse_metadata, on="Mouse ID")


# Display the data table for preview
pymaceutical_data_complete.head()
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
#drop_df = pymaceutical_data_complete.drop_duplicates(keep='last')
#drop_df

drop_df = pymaceutical_data_complete.loc[(pymaceutical_data_complete["Mouse ID"] !="g989") , :]
drop_df
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

tumor_mean = drop_df.groupby(drop_df["Drug Regimen"]).mean()["Tumor Volume (mm3)"]
tumor_median = drop_df.groupby(drop_df["Drug Regimen"]).median()["Tumor Volume (mm3)"]
tumor_variance = drop_df.groupby(drop_df["Drug Regimen"]).var()["Tumor Volume (mm3)"]
tumor_std = drop_df.groupby(drop_df["Drug Regimen"]).std()["Tumor Volume (mm3)"]
tumor_sem = drop_df.groupby(drop_df["Drug Regimen"]).sem()["Tumor Volume (mm3)"]

statistics_table = pd.DataFrame({"Mean Tumor Volume": tumor_mean,
                                 "Median Tumor Volume": tumor_median,
                                 "Tumor Volume Variance": tumor_variance,
                                 "Tumor Volume Std. Dev": tumor_std,
                                 "Tumor Volume Std. Err": tumor_sem                                                 
                                                                 
                             })

statistics_table
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.

agg_mean = drop_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].aggregate('mean')
agg_median = drop_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].aggregate('median')
agg_variance = drop_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].aggregate('var')
agg_std = drop_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].aggregate('std')
agg_sem = drop_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].aggregate('sem')

aggregate_table = pd.DataFrame({"mean": agg_mean,
                               "median": agg_median,
                               "var": agg_variance,
                               "std": agg_std,
                               "sem": agg_sem                                                 
                                                                 
                            })




aggregate_table

```
